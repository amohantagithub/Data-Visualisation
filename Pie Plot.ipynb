```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={'A':
          {'A1': 91.0, 'A2':2.4, 'A3':1.9, 'A4':5.6}, 
          'B': {'A1': 72.7, 'A2':8.8, 'A3':5.0, 'A4':13.5},
          'C': {'A1': 80.2, 'A2':6.2, 'A3':2.9, 'A4':10.7}}     
          


```

```python
data5=pd.DataFrame(data)
print(data5)
```

```python
data_P=data5.plot(kind ='pie', subplots=True, legend= None, labels=None, shadow=True, colors =['b', 'r', 'g', 'm'])
data_P[0].set_ylabel('')
data_P[0].set_title(label='A')
data_P[1].set_ylabel('')
data_P[1].set_title('B')
data_P[2].set_ylabel('')
data_P[2].set_title('C')
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.35), labels=data5.index)
plt.show()
```
